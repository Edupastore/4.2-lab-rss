### 9. Create a Pandas data frame from the feed's entries.

```python
entries = pd.DataFrame(feedburner['entries'])

entries
```

### 10. Count the number of entries per author and sort them in descending order.

```python
entries.author.value_counts().sort_values(ascending = False)
```

### 11. Add a new column to the data frame that contains the length (number of characters) of each entry title. Return a data frame that contains the title, author, and title length of each entry in descending order (longest title length at the top).

```python
n_of_characters = []
for e in entries.title:
    n_of_characters.append(len(e))
    
```

```python
entries['n_of_characters'] = n_of_characters

entries
```

```python
entries_2 = entries[['title', 'author', 'n_of_characters']]

entries_2.sort_values('n_of_characters', ascending=False)
```

### 12. Create a list of entry titles whose summary includes the phrase "machine learning."

```python
machine_learning = []
```

```python
for i in range(len(entries)):
    if 'machine learning' in entries.summary[i]:
        machine_learning.append(entries.title[i])

        
```
